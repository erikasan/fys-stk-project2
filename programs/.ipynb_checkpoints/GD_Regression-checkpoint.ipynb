```python
from sklearn.model_selection import KFold

### Create data
np.random.seed(40)

n = 20          # n x n number of data points
sigma2 = 0.01   # irreducible error
sigma = np.sqrt(sigma2)

x = np.linspace(0, 1, n)
y = np.linspace(0, 1, n)
noise = np.random.normal(0, sigma, int(n*n))
                         
# Create mesh and unravel
x, y = np.meshgrid(x, y)
x = np.ravel(x)
y = np.ravel(y)

# Observed data
z = frankeFunction(x, y) + noise

# setup for Cross-validation
max_epoch = 600
step = 50
epochs = np.arange(25, max_epoch, step, dtype=int)

# Initialize a KFold instance
k = 5
KFold = KFold(n_splits=5, shuffle=False)

scores_KFold_inv = np.zeros((len(epochs), k))
scores_KFold_mbgd1 = np.zeros((len(epochs), k))
scores_KFold_mbgd2 = np.zeros((len(epochs), k))
scores_KFold_mbgd3 = np.zeros((len(epochs), k))

deg = 5
# Generate design matrix
X = designMatrix(x, y, deg, with_intercept=False)

gamma = 0.9
seed = 42
lmbda = 0 # OLS

for i, epoch in enumerate(epochs):

    j = 0
    for train_inds, test_inds in KFold.split(x):
        X_train = X[train_inds]
        z_train = z[train_inds]

        X_test = X[test_inds]
        z_test = z[test_inds]
        
        # Scale the data, i.e, subtract the mean and divide by std (based on the training set)
        X_train_scaled, X_test_scaled = normalize_data(X_train, X_test, with_intercept=False)
        z_train_scaled, z_test_scaled = normalize_target(z_train, z_test)
        
        inv_reg = GDRegressor(fit_intercept=True)
        mbgd_reg1 = GDRegressor(lmbda=lmbda, eta=0.01, n_epochs=epoch, gamma=gamma, fit_intercept=True)
        mbgd_reg2 = GDRegressor(lmbda=lmbda, eta=0.001, n_epochs=epoch, gamma=gamma, fit_intercept=True)
        mbgd_reg3 = GDRegressor(lmbda=lmbda, eta=0.0001, n_epochs=epoch, gamma=gamma, fit_intercept=True)

        inv_reg.fit(X_train_scaled, z_train_scaled, seed_val=seed, method="Inv")
        mbgd_reg1.fit(X_train_scaled, z_train_scaled, seed_val=seed, method="MBGD")
        mbgd_reg2.fit(X_train_scaled, z_train_scaled, seed_val=seed, method="MBGD")
        mbgd_reg3.fit(X_train_scaled, z_train_scaled, seed_val=seed, method="MBGD")

        scores_KFold_inv[i, j] = inv_reg.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd1[i, j] = mbgd_reg1.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd2[i, j] = mbgd_reg2.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd3[i, j] = mbgd_reg3.mse(X_test_scaled, z_test_scaled)
        
    j += 1
    

mse_KFold_inv = np.mean(scores_KFold_inv, axis=1)
mse_KFold_mbgd1 = np.mean(scores_KFold_mbgd1, axis=1)
mse_KFold_mbgd2 = np.mean(scores_KFold_mbgd2, axis=1)
mse_KFold_mbgd3 = np.mean(scores_KFold_mbgd3, axis=1)


plt.style.use('seaborn')
fig = plt.figure(figsize=(8, 6))
plt.grid(alpha=0.3)
plt.plot(epochs, mse_KFold_inv, label='Analytical OLS')
plt.plot(epochs, mse_KFold_mbgd1, 'o--', label="Learning Rate = 0.01")
plt.plot(epochs, mse_KFold_mbgd2, 'o--', label="Learning Rate = 0.001")
plt.plot(epochs, mse_KFold_mbgd3, 'o--', label="Learning Rate = 0.0001")
plt.legend()
plt.xlabel("Number of Epochs")
plt.ylabel("MSE")
plt.title('OLS with MBGD and Various Learning Rates')
plt.show()
fig.savefig(fig_path("OLS_MSE_epochs_various_lr.pdf"), dpi=300)
```

```python
from sklearn.model_selection import KFold

### Create data
np.random.seed(40)

n = 20          # n x n number of data points
sigma2 = 0.01   # irreducible error
sigma = np.sqrt(sigma2)

x = np.linspace(0, 1, n)
y = np.linspace(0, 1, n)
noise = np.random.normal(0, sigma, int(n*n))
                         
# Create mesh and unravel
x, y = np.meshgrid(x, y)
x = np.ravel(x)
y = np.ravel(y)

# Observed data
z = frankeFunction(x, y) + noise

# setup for Cross-validation
max_epoch = 600
step = 50
epochs = np.arange(25, max_epoch, step, dtype=int)

# Initialize a KFold instance
k = 5
KFold = KFold(n_splits=5, shuffle=False)

scores_KFold_inv = np.zeros((len(epochs), k))
scores_KFold_mbgd1 = np.zeros((len(epochs), k))
scores_KFold_mbgd2 = np.zeros((len(epochs), k))
scores_KFold_mbgd3 = np.zeros((len(epochs), k))
scores_KFold_mbgd4 = np.zeros((len(epochs), k))

deg = 5
# Generate design matrix
X = designMatrix(x, y, deg, with_intercept=False)

eta = 0.001
gamma = 0.9

for i, epoch in enumerate(epochs):

    j = 0
    for train_inds, test_inds in KFold.split(x):
        X_train = X[train_inds]
        z_train = z[train_inds]

        X_test = X[test_inds]
        z_test = z[test_inds]
        
        # Scale the data, i.e, subtract the mean and divide by std (based on the training set)
        X_train_scaled, X_test_scaled = normalize_data(X_train, X_test, with_intercept=False)
        z_train_scaled, z_test_scaled = normalize_target(z_train, z_test)
        
        inv_reg = GDRegressor(fit_intercept=True)
        mbgd_reg1 = GDRegressor(eta=eta, n_epochs=epoch, gamma=gamma, batch_size=2, fit_intercept=True)
        mbgd_reg2 = GDRegressor(eta=eta, n_epochs=epoch, gamma=gamma, batch_size=8, fit_intercept=True)
        mbgd_reg3 = GDRegressor(eta=eta, n_epochs=epoch, gamma=gamma, batch_size=16, fit_intercept=True)
        mbgd_reg4 = GDRegressor(eta=eta, n_epochs=epoch, gamma=gamma, batch_size=32, fit_intercept=True)

        inv_reg.fit(X_train_scaled, z_train_scaled, seed_val=42, method="Inv")
        mbgd_reg1.fit(X_train_scaled, z_train_scaled, seed_val=42, method="MBGD")
        mbgd_reg2.fit(X_train_scaled, z_train_scaled, seed_val=42, method="MBGD")
        mbgd_reg3.fit(X_train_scaled, z_train_scaled, seed_val=42, method="MBGD")
        mbgd_reg4.fit(X_train_scaled, z_train_scaled, seed_val=42, method="MBGD")

        scores_KFold_inv[i, j] = inv_reg.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd1[i, j] = mbgd_reg1.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd2[i, j] = mbgd_reg2.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd3[i, j] = mbgd_reg3.mse(X_test_scaled, z_test_scaled)
        scores_KFold_mbgd4[i, j] = mbgd_reg4.mse(X_test_scaled, z_test_scaled)
        
    j += 1
    

mse_KFold_inv = np.mean(scores_KFold_inv, axis=1)
mse_KFold_mbgd1 = np.mean(scores_KFold_mbgd1, axis=1)
mse_KFold_mbgd2 = np.mean(scores_KFold_mbgd2, axis=1)
mse_KFold_mbgd3 = np.mean(scores_KFold_mbgd3, axis=1)
mse_KFold_mbgd4 = np.mean(scores_KFold_mbgd4, axis=1)


plt.style.use('seaborn')
fig = plt.figure(figsize=(8, 6))
plt.grid(alpha=0.3)
plt.plot(epochs, mse_KFold_inv, label='Analytical OLS')
plt.plot(epochs, mse_KFold_mbgd1, 'o--', label="Batch size = 2")
plt.plot(epochs, mse_KFold_mbgd2, 'o--', label="Batch size = 8")
plt.plot(epochs, mse_KFold_mbgd3, 'o--', label="Batch size = 16")
plt.plot(epochs, mse_KFold_mbgd4, 'o--', label="Batch size = 32")
plt.legend()
plt.xlabel("Number of Epochs")
plt.ylabel("MSE")
plt.title('OLS MBGD With Batch Size')
plt.show()
fig.savefig(fig_path("OLS_MSE_epochs_particular_batch_size.pdf"), dpi=300)
```
